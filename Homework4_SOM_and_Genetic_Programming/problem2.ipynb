```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fastsr.estimators.symbolic_regression import SymbolicRegression
from fastgp.algorithms.fast_evaluate import fast_numpy_evaluate
from fastgp.parametrized.simple_parametrized_terminals import get_node_semantics
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
data = pd.read_csv("thermal-dataset.csv")
#print(data)
target = data['y-exp'].values

data = data[data.columns[1:21]]

# data normalization
data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
data = data.values
```

```python
# def target(x):
#     return x**3 + x**2 + x

# X = np.linspace(-10, 10, 100, endpoint=True)
# y = target(X)
# X_train, X_test, y_train, y_test = train_test_split(X, y)

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
sr = SymbolicRegression(ngen = 60, pop_size = 100)
sr.fit(X_train, y_train)
score = sr.score(X_test, y_test)
print('Score: {}'.format(score))
```

```python
print('Best Individuals:')
sr.print_best_individuals()
```

```python
history = sr.history_
population = list(filter(lambda x: hasattr(x, 'error'), list(sr.history_.genealogy_history.values())))
population.sort(key=lambda x: x.error, reverse=True)
```

```python
#data = data.reshape((len(data), 1))
i = 1
previous_errror = population[0]
unique_individuals = []
while i < len(population):
    ind = population[i]
    if ind.error != previous_errror:
        print(str(i) + ' | ' + str(ind.error) + ' | ' + str(ind))
        unique_individuals.append(ind)
    previous_errror = ind.error
    i += 1
```

```python
def plot(index):
    plt.plot(data, target, 'r')
    plt.axis([-10, 10, -1000, 1000])
    y_hat = fast_numpy_evaluate(unique_individuals[index], sr.pset_.context, data, get_node_semantics)
    plt.plot(data, y_hat, 'g')
    plt.savefig('symbolic_regression_plots/' + str(i) + 'ind.png')
    plt.gcf().clear()

i = 0
while i < len(unique_individuals):
    plot(i)
    i += 10
i = len(unique_individuals) - 1
plot(i)
```

```python
y_predict = sr.predict(X_test)
```

```python
fig = plt.figure()
sns.set_theme(color_codes=True)
ax = sns.regplot(x=y_test, y=y_predict, line_kws={"color": "black"})

plt.xlim(-5, 200)
plt.xlabel('targets')
plt.ylabel('predictions')
plt.show()
fig.savefig('scatter-plot.png')
```
